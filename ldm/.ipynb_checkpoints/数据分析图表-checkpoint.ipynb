```python
import mysql.connector
from sqlalchemy import create_engine
import pandas as pd
from pyecharts.charts import Bar
from pyecharts import options as opts

# 创建 SQLAlchemy 引擎
engine = create_engine('mysql+mysqlconnector://root:123456@localhost/students01')

# 查询数据
query = """
SELECT 
    年龄, 
    性别, 
    AVG(学业压力) AS avg_pressure
FROM students_02
GROUP BY 年龄, 性别
ORDER BY avg_pressure ASC
"""
df = pd.read_sql(query, engine)

# 创建柱状图
bar = Bar()
bar.add_xaxis(df['年龄'].astype(str).tolist())
bar.add_yaxis("男性平均学业压力", df[df['性别'] == '男']['avg_pressure'].tolist())
bar.add_yaxis("女性平均学业压力", df[df['性别'] == '女']['avg_pressure'].tolist())

# 设置全局配置
bar.set_global_opts(
    title_opts=opts.TitleOpts(title="不同年龄不同性别的平均学业压力"),
    xaxis_opts=opts.AxisOpts(name="年龄"),
    yaxis_opts=opts.AxisOpts(name="平均学业压力"),
    legend_opts=opts.LegendOpts(pos_left="right")
)

# 在 Jupyter Notebook 中显示图表
bar.render_notebook()
```
